```python
def drawImg(labels,colors):
    
    labels = labels.reshape((x,y))
    new_img = np.zeros((x,y,3),dtype='uint8')
    for i in range(x):
        for j in range(y):
            new_img[i,j,:] = colors[labels[i][j]]
    new_img = new_img.reshape((original_shape))
    plt.axis("off")
    plt.imshow(new_img)
    plt.show()
```
