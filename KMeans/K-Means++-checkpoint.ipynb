#### SCIKIT Learn by default uses the K-means++ approach to find centers.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

# Using SciKit Learn

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
print(kmeans)
```

```python
kmeans.fit(X,Y) # to train model
```

```python
centers = kmeans.cluster_centers_ # to get centers
print(centers)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(centers[:,0],centers[:,1],marker='*',color='red')
plt.title('Prediction')
plt.show()
```

```python
ypred = kmeans.labels_
```

```python
print(ypred.shape, Y.shape)
```

```python
print(np.unique(Y, return_counts = True))
print(np.unique(ypred, return_counts=True))
```
